```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
all_ce_offdiag_melt = pd.read_csv("dataset/orthomeasures/ce_offdiag_measures.csv")
all_opl_offdiag_melt = pd.read_csv("dataset/orthomeasures/opl_offdiag_measures.csv")
all_qr_offdiag_melt = pd.read_csv("dataset/orthomeasures/qr_offdiag_measures.csv")
```

```python
# seaborn line plot
plt.figure(figsize=(3., 3.2))
# y labl
# lowerarrow = r'$\downarrow$'
arrow_down = '\u2193'
plt.ylabel(f'Ortho Measure({arrow_down})', fontsize=16, fontweight='bold')
# light grey grid

plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)

linewidth = 3
sns.lineplot(data=all_ce_offdiag_melt, x='epoch', y='offdiag', label='CE only', color='red', linewidth=linewidth)
sns.lineplot(data=all_opl_offdiag_melt, x='epoch', y='offdiag', label='CE+OPL', color='green', linewidth=linewidth)
label_notation = f""
sns.lineplot(data=all_qr_offdiag_melt, x='epoch', y='offdiag', label=r'CE+QR ($\lambda$=0.5)', color='blue', linewidth=linewidth)

# make legend thicker and bold
plt.legend(fontsize=12)
# get legend
leg = plt.gca().get_legend()
for line in leg.get_lines():
    line.set_linewidth(5)

#
# x ticks
plt.xticks([0, 5, 10 , 15, 20], fontsize=14)
# y ticks
plt.yticks([0.05, 0.10, 0.15, 0.20, 0.25], fontsize=14)
# x label
plt.xlabel('Epoch', fontsize=16, fontweight='bold')
# save
plt.savefig('results/orthogonality/ortho_qr_measure.png', bbox_inches='tight', dpi=300)
```
